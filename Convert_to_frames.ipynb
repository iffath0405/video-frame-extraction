```python
import cv2
import os

video_path = "Video.mp4"
out_dir = "frames"
os.makedirs(out_dir, exist_ok=True)

cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print("❌ Video is not loaded, Please check your video path")
    exit()

fps = cap.get(cv2.CAP_PROP_FPS)
frame_interval = int(round(fps))  # 1 second = fps frames

count = 0      # all frames count
save_count = 0 # saved frames count

while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        break  # video finished / read failed

    if count % frame_interval == 0:
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        ok = cv2.imwrite(f"{out_dir}/frame_{save_count:04d}.jpg", gray)
        if not ok:
            print("❌ Save fail:", save_count)
        save_count += 1

    count += 1

cap.release()
print("FPS :" ,fps)
print("✅ Done. Saved frames:", save_count)
```
